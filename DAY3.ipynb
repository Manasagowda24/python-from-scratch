```python
''' Reverse a number using for loop '''
n = int(input("Enter number: "))
rev = 0

for i in range(len(str(n))):
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10

print("Reverse:", rev)
```

```python
'''Check palindrome number '''
n = int(input("Enter number: "))
temp = n
rev = 0

for i in range(len(str(n))):
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10

if temp == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
## Check if a number is prime
n = int(input("Enter a number: "))

i = 2
is_prime = True

while i < n:
    if n % i == 0:
        is_prime = False
        break
    i += 1

if n > 1 and is_prime:
    print("Prime")
else:
    print("Not Prime")
```

```python
''' Number guessing game:
Keep asking input
Stop when correct
Skip invalid inputs '''


secret = 7   # correct number

while True:
    n = input("Enter a number: ")

    # check invalid input
    if not n.isdigit():
        print("Invalid input, try again")
        continue

    n = int(n)

    if n == secret:
        print("Correct! 🎉")
        break
    else:
        print("Wrong, try again")
```
